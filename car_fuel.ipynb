Importing required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as stat 

# Load the data
data = pd.read_csv('CAR FUEL DATA SET.csv')
```

Plotting Histogram for getting information about Engine Displacement

```python
# 1. Histogram of Engine Displacement
plt.figure(figsize=(7, 5))
plt.hist(data['Engine Displacement'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Engine Displacement')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Plotting Scatter plot for City fuel economy vs Highway fuel economy

```python
# 2. Scatter Plot of City Fuel Economy vs Highway Fuel Economy
plt.figure(figsize=(7, 5))
plt.scatter(data['City FE'], data['Highway FE'], c='red', alpha=0.5)
plt.title('City Fuel Economy vs Highway Fuel Economy')
plt.xlabel('City Fuel Economy (mpg)')
plt.ylabel('Highway Fuel Economy (mpg)')
plt.grid(True)
plt.show()
```

Getting information abou average combined fuel economy by manufacturer

```python
# 3. Bar Chart of Average Combined Fuel Economy by Manufacturer
avg_combined_fe = data.groupby('Mfr Name')['Combined FE'].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_combined_fe.plot(kind='bar', color='green')
plt.title('Average Combined Fuel Economy by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Combined Fuel Economy (mpg)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

Box Plot for CO2 emission

```python
# 4. Box Plot of CO2 Emissions
plt.figure(figsize=(7, 5))
plt.boxplot([data['City CO2'], data['Highway CO2'], 
             data['Combined CO2']], 
            labels=['City CO2', 'Highway CO2', 'Combined CO2'])
plt.title('CO2 Emissions Distribution')
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(True)
plt.show()
```

Analysing the city fuel economy by plotting pie chart

```python
# Define bins for City Fuel Economy
bins = [0, 10, 20, 30, 40, 50, 60]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60']

# Categorize City Fuel Economy into bins
data['City FE Category'] = pd.cut(data['City FE'], bins=bins, labels=labels)

# Calculate counts for each category
city_fe_counts = data['City FE Category'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(city_fe_counts, labels=city_fe_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of City Fuel Economy Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend()
plt.show()
```
