# Analysis of Age Distribution for Individuals with Positive Diabetes Outcome

```python
#for individual ages vs diabetes positive
#getting data of patients who has positive diabetes outcome
positive_diabetes = data[data['Outcome'] == 1]
age_distribution = positive_diabetes['Age'].value_counts()  #getting age counts

#plotting pie chart
plt.figure(figsize=(10, 25))
age_distribution.plot(kind='pie', autopct='%1.1f%%',startangle = 180  )
plt.title('Age Distribution of Individuals with Positive Diabetes Outcome') 
plt.show()
```

Analysis of positive diabetes patients with age distribution Age above 50 (age > 50) and age below 50(age <= 50) categories

```python
#pie chart for age distribution for positive diabetes patients
positive_diabetes = data[data['Outcome'] == 1]

#slicing the age data 
#storing positive diabetes patients data who all are above age 50
age_group_over_50 = positive_diabetes[positive_diabetes['Age'] > 50].shape[0] 
#storing the positive diabetes patients below age 50
age_group_50_or_less = positive_diabetes[positive_diabetes['Age'] <= 50].shape[0]

#storing both data in a single list 
age_groups = ['Age > 50', 'Age <= 50']
counts = [age_group_over_50, age_group_50_or_less]

#plotting
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=age_groups, autopct='%1.1f%%', startangle=140, colors=['gray', 'rosybrown'])#plotting pie chart for age groups
plt.title('Age Distribution of Individuals with Positive Diabetes Outcome')
plt.legend()
plt.show()
```

```python
from scipy.stats import skew,kurtosis
sk_age = skew(age)
sk_bmi = skew(bmi)
sk_bp = skew(bpressure)
sk_gl = skew(glucose)
print(sk_age,sk_bmi,sk_bp,sk_gl)
print("Skewness :")
def skewness(skew_value, attr):
    if skew_value == 0:
        print(f"{attr} is normally skewed")
    elif skew_value > 0:
        print(f"{attr} has positive skew")
    elif skew_value < 0:
        print(f"{attr} has negative skew")

#print(sk_age, sk_bmi, sk_bp, sk_gl)

skewness(sk_age, "Age")
skewness(sk_bmi, "BMI")
skewness(sk_bp, "Blood Pressure")
skewness(sk_gl, "Glucose")

k_age = kurtosis(age)
k_bmi = kurtosis(bmi)
k_bp = kurtosis(bpressure)
k_gl = kurtosis(glucose)
print("Kurtosis :\n","Age :",k_age,"\n","BMI :",k_bmi,"\n","Blood Pressure :",k_bp,"\n","Glucose :",k_gl)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create the boxplot
plt.boxplot(data, 
            notch=True, 
            vert=True, 
            patch_artist=True, 
            labels=['Group 1', 'Group 2', 'Group 3'],
            showmeans=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='red', linewidth=2),
            capprops=dict(color='black', linewidth=2),
            medianprops=dict(color='green', linewidth=2),
            meanprops=dict(marker='D', markeredgecolor='black', markerfacecolor='red'))

# Add title and labels
plt.title('Customized Boxplot')
plt.xlabel('Groups')
plt.ylabel('Values')

# Show the plot
plt.show()
```
