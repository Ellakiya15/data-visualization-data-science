Analysing Amazon and Apple stock csv to get useful data

Importing required Libraries for plotting as well Normalizing and standardizing the data

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, FunctionTransformer

amazon = pd.read_csv('Amazon Stock.csv')
apple = pd.read_csv("Apple Stock.csv")

azv = amazon[['Volume']]
apv = apple[['Volume']]

scaler = MinMaxScaler()
azv1 = scaler.fit_transform(azv)
apv1 = scaler.fit_transform(apv)
```

Getting information of both amazon and apple volume using overlapping

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
vol1 = plt.plot(amazon.Volume , label = 'Amazon Volume')
vol2 = plt.plot(apple.Volume, label = 'Apple Volume')
plt.title(' Volume without Normalizing')
plt.legend()
plt.subplot(1,2,2)
plt.plot(azv1, label='Amazon volume')
plt.plot(apv1, label='Apple volume')
plt.title('Volume with Normalizing')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
highaz = amazon[['High']]
highap = apple[['High']]
date = amazon[['Date']]

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(highaz, label= 'High in Amazon')
plt.plot(highap, label= 'High in Apple')
plt.legend()
plt.subplot(1,2,2)
plt.hist(highaz,color='Blue',alpha = 0.7,label='Amazon High')
plt.hist(highap,color='Red',alpha = 1,label='Apple High')
plt.legend()
plt.tight_layout()
plt.show()
```
