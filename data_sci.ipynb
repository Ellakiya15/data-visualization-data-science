Scatter plot for CO2 vs Volume

```python
# Scatter plot: CO2 Emissions vs. Engine Volume
plt.figure(figsize=(10, 6))
plt.scatter(data['Volume'], data['CO2'])
plt.title('CO2 Emissions vs. Engine Volume')
plt.xlabel('Engine Volume (cc)')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```

Bar plot to analyse the CO2 emission for each car model

```python
# Bar plot: Average CO2 Emissions by Car Model
plt.figure(figsize=(5, 4))
avg_co2 = data.groupby('Model')['CO2'].mean().sort_values()
avg_co2.plot(kind='bar')
plt.title('Average CO2 Emissions by Car Model')
plt.xlabel('Car Model')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.xticks(rotation=90)
plt.show()
```

 Box plot: Distribution of CO2 Emissions by Car Manufacturer

```python
# Box plot: Distribution of CO2 Emissions by Car Manufacturer
plt.figure(figsize=(14, 8))
data.boxplot(column='CO2', by='Car', grid=False)
plt.title('Distribution of CO2 Emissions by Car Manufacturer')
plt.suptitle('')
plt.xlabel('Car Manufacturer')
plt.ylabel('CO2 Emissions (g/km)')
plt.xticks(rotation=90)
plt.show()
```

Histogram: Distribution of CO2 Emissions

```python
# Histogram: Distribution of CO2 Emissions
plt.figure(figsize=(10, 6))
plt.hist(data['CO2'], bins=15,edgecolor='black',label = 'CO2')
plt.legend()
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()
```
